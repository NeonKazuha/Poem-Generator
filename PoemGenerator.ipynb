```python
import numpy as np
import string
import random
```

```python
input_text = []
for line in open('robert_frost.txt'):
    line = line.rstrip().lower()
    if line:
      line = line.translate(str.maketrans('', '', string.punctuation)) # Removes all Punctuations from the line
      input_text.append(line)
```

```python
input_text
```

```python
pi_dictionary = {}
for line_number, line in enumerate(input_text): # Enumerate gives (Line_Number, "Line_text")
  tokens = line.split()
  for token in tokens:
    if token in pi_dictionary:
      pi_dictionary[token] += 1
    else:
      pi_dictionary[token] = 1
    break
```

```python
total_first_words = len(input_text)
for token in pi_dictionary:
  pi_dictionary[token] /= total_first_words
```

```python
first_order_A = {}
second_order_A = {}
for line in input_text: # Did not use enumerate so didn't need to specify line number
  tokens = line.split() # A list of tokens is created, that means, every token in tokens has an index to call it
  for i in range(len(tokens)):
    if i == 1:
      first_word = tokens[i-1]
      second_word = tokens[i]
      if first_word not in first_order_A:
        first_order_A[first_word] = {}
      if second_word not in first_order_A[first_word]:
          first_order_A[first_word][second_word] = 1
      else:
          first_order_A[first_word][second_word] += 1
    elif i > 1:
      first_word = tokens[i-2]
      second_word = tokens[i-1]
      current_word = tokens[i]
      if first_word not in second_order_A:
        second_order_A[first_word] = {}
      if second_word not in second_order_A[first_word]:
        second_order_A[first_word][second_word] = {}
      if current_word not in second_order_A[first_word][second_word]:
        second_order_A[first_word][second_word][current_word] = 1
      else:
        second_order_A[first_word][second_word][current_word] += 1
```

```python
def first_order_probabilities(dictionary):
    probs_foa = {}
    for first_word, second_word_count in dictionary.items():
        total_count = sum(second_word_count.values())
        probs_foa[first_word] = {}
        for second_word, count in second_word_count.items():
            probability = count / total_count
            probs_foa[first_word][second_word] = probability
    return probs_foa
```

```python
bigram_probs = first_order_probabilities(first_order_A)
```

```python
def sample_word(dictionary):
    p0 = np.random.random()
    cumulative = 0
    for word, probability in dictionary.items():
        cumulative += probability
        if p0 < cumulative:
            return word
    assert False
```

```python
def generate():
    sentences = []

    first_word = sample_word(pi_dictionary)
    sentences.append(first_word)

    second_word = sample_word(bigram_probs[first_word])
    sentences.append(second_word)

    for j in range(8):
      third_word = sample_word(trigram_probs[first_word][second_word])
      sentences.append(third_word)
      first_word = second_word
      second_word = third_word
    print(' '.join(sentences))
```
